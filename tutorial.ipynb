## visualize the results saved in the json in above evaluation part 

````python
import json
result_json = open('/home/lvsolo/Downloads/hard-hat-detection/codes/yolov5/Dataset/annotations/instances_val.json', 'r')
js = json.load(result_json)
print(js.keys())
print(js['categories'][0])
print(js['annotations'][0])
print(js['images'][0])
for k in js.keys():
    print(len(js[k]))
# js['images']
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
images_dir = '/home/lvsolo/Downloads/hard-hat-detection/codes/yolov5/Dataset/images/val/'

# print(ga.keys())
image_name = ""
for ga in js['annotations']:
    print(ga['bbox'])
    print(ga['image_id'])
    if image_name == "":
        image_name = ga['image_id']
        image_path = images_dir + '/' + ga['image_id'] + '.png'
        image = cv2.imread(image_path)
    elif image_name != ga['image_id']:
        cv2.imwrite('runs/val/'+image_name+'.png', image)
        # break
        image_name = ga['image_id']
        image_path = images_dir + '/' + ga['image_id'] + '.png'
        image = cv2.imread(image_path)
    x0 , y0, x1, y1 = ga['bbox']
    x1 = int(x1 + x0)
    y1 = int(y1 + y0)
    x0 = int(x0)
    y0 = int(y0)
    image = cv2.rectangle(image, (x0,y0), (x1,y1), (255,0,0),2)
# plt.imshow(image)
````
